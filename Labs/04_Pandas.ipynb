## 4.3 Basic Dataframe Operations

This section will go through combining dataframes and basic slices and dices.

```python
df_canada = pd.concat([rgi_03,rgi_04])#stack the two dataframes on top of each other
```

#### This sub-section goes over subsetting by index and by location (iloc/loc)

```python
df_canada_sub1 = df_canada.iloc[0:5]#Get the first 5 rows and put into a new dataframe
df_canada_sub1.head(14)
```

```python
df_canada_sub1 = df_canada.iloc[20:25]#Get the rows 20-25 and put into a new dataframe
df_canada_sub1.head(14)
```

```python
df_canada_sub3 = df_canada.loc[:, ['RGIId']]## Selects the RGIId column and puts in new dataframe. The colon gets all rows
```

```python
df_canada_sub4 = df_canada.loc[:, ['RGIId','GLIMSId']]## Selects the RGIId column and GLIMSID column and puts in new dataframe
df_canada_sub4.head()
```

```python
df_canada_sub5 = df_canada.loc[[0, 2]]
df_canada_sub5.head()### What is going on here??? Why are there 4 rows instead of 2?
```

```python
df_canada = df_canada.reset_index()
```

```python
df_canada_sub5 = df_canada.loc[[0,2]]
df_canada_sub5.head()
```

```python
df_canada_sub6 = df_canada.loc[[0,2],["Zmin","Zmed","Zmax"]]#Slice by row value and column value
df_canada_sub6.head()
```

### Selecting by values

```python
df_canada_sub7 = df_canada[df_canada['Zmin'] < 600]#Select all glaciers with minimum elevation less than 600m
df_canada_sub7.head()
```

```python
df_canada_sub8 = df_canada[df_canada['Zmin'] > 1000]#Select all glaciers with minimum elevation greater than 1000m
df_canada_sub8.head()
```

```python
df_canada_sub9 = df_canada[df_canada['Zmin'] == 1000]#Select all glaciers with minimum elevation exactly equal to 1000m
df_canada_sub9.head()
```

You can make multiple slices too using conditional statements. & means and and | means or

```python
df_canada_sub10 = df_canada[(df_canada['Zmin'] == 1000) & (df_canada['Area'] < 2)]#Select all glaciers with minimum elevation exactly equal to 1000m
df_canada_sub10.head()
```

### 4.4 Writing to a file

You can write your csv to a flat file using to_csv. This is helpful for doing operations using python then exporting it to a file

```python
df_canada_sub10.to_csv('df_canada_sub.csv',index = False)#Index=False makes sure there isn't a new index column
```
