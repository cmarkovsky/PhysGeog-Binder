### 2.2 - Using `pandas` to load tabular data

`pandas` has its own datatype that it uses for tabular data called a `DataFrame`. We can load in data from many different file types (.csv, .xlsx), but we'll just go through a simple example here.

```python
path_to_data = '../../data/planetary_climates.csv'

df = pd.read_csv(path_to_data)
df
```

If we want to access a subset of this table, we can index by either row(s) or column(s).

```python
### Index by row. .iloc will given use the integer location (7 on the table above). Note that Python starts indexing at 0.
measured_temp = df.iloc[7]
print(measured_temp)
```

```python
### Index by column
earth = df['Earth']
earth
```

### 2.3 - Plotting with `matplotlib`

Plotting is one of the most powerful tools in science. Visualization is critical in digesting infomration and plotting is extremely helpful both for communicating your findings and for sanity checks while problem solving.

We'll just do a few basic examples using the `DataFrame` we loaded earlier. There are a few data manipulation steps here that we need to do to get the data into the right format to plot (plotting usually works well with numerical data, less with other types of data).

```python
x = ['Mercury', 'Venus', 'Earth', 'Mars', 'Titan', 'Pluto']
y = list(measured_temp[3:])

plt.plot(x, y);
```

This doesn't look great, you can get some idea of the data, but if you handed this to someone else, they would have no idea what you're showing. Let's do a bit better

```python
fig, ax = plt.subplots()
ax.scatter(x,y, color = 'orange');
ax.set_xlabel('Planet');
ax.set_ylabel('Temperature (K)');
ax.set_title('Measured Planetary Temperatures');
```

Getting better, but maybe we're curious to compare the temperature with some of the other variables in our table. Let's make one more plot that let's us do a side-by-side comparison of temperature, distance to the sun, and albedo.

```python
fig, ax = plt.subplots(3, 1, sharex = True, figsize = (8,8))
x = ['Mercury', 'Venus', 'Earth', 'Mars', 'Titan', 'Pluto']

y1 = list(df.iloc[7][3:])
y2 = list(df.iloc[3][3:])
y3 = list(df.iloc[1][3:])

ax[0].scatter(x, y1, color = 'orange');
ax[1].scatter(x, y2, color = 'green');
ax[2].scatter(x, y3, color = 'blue');

ax[2].set_xlabel('Planet');

ax[0].set_ylabel('Temperature (K)', fontsize = 12);
ax[1].set_ylabel('Distance from Sun (m)', fontsize = 12);
ax[2].set_ylabel('Albedo', fontsize = 12);

fig.suptitle('Planetary Climates', fontsize = 14)
plt.tight_layout()
```
