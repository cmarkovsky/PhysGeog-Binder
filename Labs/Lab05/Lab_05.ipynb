### 5.1 Calculating PCO2 assuming variable gas transfer

$$
CO_2(x) = C \, e^{\left(-\frac{k}{w \cdot h} \, x\right)} + CO_{2,\text{atmos}}
$$

```python
C = 11580
CO2 = pd.read_csv('CO2.csv')
variable_k = [0.00022,0.00044,0.00068,0.00089]
pC02_a = 420

const_w = 6
const_H = 7
```
