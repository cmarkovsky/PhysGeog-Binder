## 7.2 - Hydrolic Gradient Calculation
We are particularly interested in the impact of sea-level rise (SLR) on these two regions. An important piece of this analysis is examining the hydraulic gradient for different magnitudes of SLR. The general form of the equation for hydraulic gradient is:

$$
G = \frac{H_1 - H_2}{L}
$$

Let's use values of SLR ranging from 0 to 2 meters: `[0, 0.25, 0.5, 0.75, 1, 1.25, 1.5, 1.75, 2]`

```python
# Make an array defining our SLR values

SLRs = [0, 0.25, 0.5, 0.75, 1, 1.25, 1.5, 1.75, 2]

df_tampa = pd.DataFrame(SLRs, columns = ['SLR'])
df_seattle = df_tampa.copy()
```

```python
# Let's make a function to calculate this.
def calc_G(H1, H2, L):
    G = (H1 - H2) / L
    return G
```

```python
Gs_tampa = []
Gs_seattle = []

H_tampa = data.loc['Tampa']['H_well']
H_seattle = data.loc['Seattle']['H_well']
L = data.loc['Tampa']['distance']

for SLR in SLRs:
    G_tampa = calc_G(H_tampa, SLR, L)
    G_seattle = calc_G(H_seattle, SLR, L)
    Gs_tampa.append(G_tampa)
    Gs_seattle.append(G_seattle)

df_tampa['G'] = Gs_tampa
df_seattle['G'] = Gs_seattle
```

## 7.3 - Wedge Shape Calculation
To calculate the shape of the salt water wedge ($a$), we can use a simple equation that depends on salt water's specific gravity ($s$), the depth of the aquifer ($d$), the hydrolic gradient ($G$), and the depth of the aquifer below sea-levl ($b$):
$$
a = \frac{s(s-1)d^2}{2Gb}
$$

```python
# Let's make a function to calculate this. 
def calc_a(d, G, b, s=spec_grav):
    a = (s * (s-1) * d**2) / (2*G*b)
    return a
```

```python
# Now, let's make a for loop to calculate the shape of each wedge for a given sea-level rise magnitude
as_tampa = []
as_seattle = []
d_tampa = data.loc['Tampa']['depth_d']
b_tampa = data.loc['Tampa']['depth_b']

d_seattle = data.loc['Seattle']['depth_d']
b_seattle = data.loc['Seattle']['depth_b']

for i in range(len(SLRs)):
    G_tampa = df_tampa['G'][i]
    a_tampa = calc_a(d_tampa, G_tampa, b_tampa)
    as_tampa.append(a_tampa)

    G_seattle = df_seattle['G'][i]
    a_seattle = calc_a(d_seattle, G_seattle, b_seattle)
    as_seattle.append(a_seattle)

df_tampa['a'] = as_tampa
df_seattle['a'] = as_seattle
```

```python
# Percent Difference

def calc_per_diff(a1, a2):
    diff = (a2 - a1) / a1
    return diff * 100
```

```python
a1_tampa = df_tampa['a'][0]
a1_seattle = df_seattle['a'][0]

diffs_tampa = []
diffs_seattle = []

for i in range(len(df_tampa['a'])):
    diff_tampa = calc_per_diff(a1_tampa, df_tampa['a'][i])
    diffs_tampa.append(diff_tampa)
    
    diff_seattle = calc_per_diff(a1_seattle, df_seattle['a'][i])
    diffs_seattle.append(diff_seattle)

df_tampa['per_diff'] = diffs_tampa
df_seattle['per_diff'] = diffs_seattle    
    
```
